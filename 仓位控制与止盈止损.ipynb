```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [6,4]
plt.rcParams['figure.dpi'] = 150
```

## 假设一共进行200次交易

```python
t = 200
n = pd.Series(np.random.randint(0,2, size=t))
```

```python
n.replace(0,-1,inplace = True)
n.head()
```

```python
sns.countplot(x=n)
plt.show()
```

## 不设仓位控制
## 不加止盈止损
## 全部“一把梭”

```python
x1=[200000]
for i in range(t):
    a = x1[-1]*(1+n[i])
    x1.append(a)
print(x1[-1])
```

```python
pd.Series(x1).plot(title= 'Total Asset')
plt.show()
```

## 接下来，我们看一下
## 每次只用25%仓位
## 止盈止损全部为10%

```python
x2 = [200000]
for i in range(t):
    a = x2[-1]*0.25*(1+n[i]*0.1)+x2[-1]*0.75
    x2.append(a)
print(x2[-1])
```

```python
pd.Series(x2).plot(title= 'Total Asset')
plt.show()
```

## 接下来
## 每次交易仓位控制在10%
## 20%止盈，10%止损

```python
x3 = [250000]
for i in range(t):
    if n[i] == 1:
        a = x3[-1]*0.025*(1+n[i]*0.2) + x3[-1]*0.975
    else:
        a = x3[-1]*0.025*(1+n[i]*0.1) + x3[-1]*0.975
    x3.append(a)
print(x3[-1])
```

```python
pd.Series(x3).plot(title= 'Total Asset')
plt.show()
```

# 结论：
## 合理的仓位控制与止盈止损
## 才能让我们立于不败之地
